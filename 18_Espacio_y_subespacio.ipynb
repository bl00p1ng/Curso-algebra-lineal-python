# Clase 18. ¿Qué es un espacio y un subespacio?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v_1 = np.array([1, 1])
v_2 = np.array([-1, -1])
```

```python
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v_1[0] * a + v_2[0] * b,
                    v_1[1] * a + v_2[1] * b,
                    marker='.', color = 'orange')
        
plt.xlim(-25, 25)
plt.ylim(-25, 25)

plt.axvline(x=0, color='gray')
plt.axhline(y=0, color='gray')

plt.show()
```

```python
v_1 = np.array([1, 0])
v_2 = np.array([2, -3])
```

```python
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v_1[0] * a + v_2[0] * b,
                    v_1[1] * a + v_2[1] * b,
                    marker='.', color = 'orange')
        
plt.xlim(-25, 25)
plt.ylim(-25, 25)

plt.axvline(x=0, color='gray')
plt.axhline(y=0, color='gray')

plt.show()
```

```python
from mpl_toolkits import mplot3d

v_1 = np.array([1, 0, 0])
v_2 = np.array([2, -3, 0])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

for a in range(-10, 10):
    for b in range(-10, 10):
        ax.scatter(v_1[0] * a + v_2[0] * b,
                    v_1[1] * a + v_2[1] * b,
                    v_1[2] * a + v_2[2] * b,
                    marker='.', color = 'orange')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```
